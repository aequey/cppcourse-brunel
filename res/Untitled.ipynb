```python
import numpy as np
import pylab as pl

data = np.genfromtxt('data2.dat')
```

```python
select= np.array ([d for d in data if d[1] < 50])
data1= select.transpose()

pl.scatter(data1[0],data1[1],alpha=0.8, edgecolors='none');
pl.show();
```

```python
select= np.array ([d for d in data])
data1= select.transpose()
pl.hist(data1, 1);
pl.show();
```
